```python
def plot_pareto_01(f1,pla,ax,title=None,color=False,y=None,NOAX = False,lim=None):
    for i in range(len(f1)):
        color_expl = colors[i]
        for j in range(len(f1[0])):
            color_gnn = colors[j]

            if lim == 0.5:
                alpha = 0.55
            else:
                alpha = 0.05
                
            if f1[i][j] >= alpha and pla[i][j]>= alpha:
               
            #if True: 
                if color:
                    if y==1:
                        ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN_bahousescolor[j]+"("+str(1)+")",fontdict=dict(color=color_expl,size=10),weight="bold")
                        #ax.scatter(f1[i][j],pla[i][j])
                    else:
                        ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN_bahousescolor[j]+"("+str(0)+")",fontdict=dict(color=color_expl,size=10),weight="bold")
                else:
                    ax.text(x=f1[i][j]+0.01,y=pla[i][j]-0.02,s=GNN[j],fontdict=dict(color=color_expl,size=10),weight="bold")
                    #ax.scatter(f1[i][j],pla[i][j],color=color_expl,s=10)
                        
                #ax.text(x=f1[i][j],y=pla[i][j],s=expl[i],fontdict=dict(color=color_gnn,size=10),weight="bold")


    ax.set_ylabel("Plausibility")
    ax.set_xlabel("F1-Fidelity")
    
    if lim == 0.5:
        ax.set_xlim(0.5,1)
        ax.set_ylim(0.5,1)
    else:
        ax.set_xlim(0,1)
        ax.set_ylim(0,1)
        
    
    if not title == None:
        ax.set_title(title)


    legend_elements_expl = []
    for i in range(len(expl)):
        ax.scatter(-1,-1,color=colors[i],label=expl[i])
    for i in range(len(expl)):
        legend_elements_expl.append(Line2D([0], [0], marker='o', color=colors[i], label=expl[i],))

    return legend_elements_expl
```

```python

def plot_pareto_tmp(f1,pla,ax,title=None,color=False,y=None):
    legend = True
    for i in range(len(f1)):
        color_expl = colors[i]
        for j in range(len(f1[0])):
            color_gnn = colors[j]

            if f1[i][j] >= 0.75 and pla[i][j]>= 0.75:
               
            #if True: 
                if color:
                    if y==1:
                        ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN_bahousescolor[j]+"("+str(1)+")",fontdict=dict(color=color_expl,size=10),weight="bold")
                    else:
                        legend = False
                        ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN_bahousescolor[j]+"("+str(0)+")",fontdict=dict(color=color_expl,size=10),weight="bold")
                        
                else:
                    ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN[j],fontdict=dict(color=color_expl,size=10),weight="bold")
                #ax.text(x=f1[i][j],y=pla[i][j],s=expl[i],fontdict=dict(color=color_gnn,size=10),weight="bold")


    ax.set_ylabel("Plausibility")
    ax.set_xlabel("Fidelity")

    ax.set_xlim(0.7,1)
    ax.set_ylim(0.7,1)
    #ax.set_xlim(0,1)
    #ax.set_ylim(0,1)
    
    if not title == None:
        ax.set_title(title)

    if legend:
        legend_elements_expl = []
        for i in range(len(expl)):
            ax.scatter(-1,-1,color=colors[i],label=expl[i])
        for i in range(len(expl)):
            legend_elements_expl.append(Line2D([0], [0], marker='o', color=colors[i], label=expl[i],))
    
        return legend_elements_expl

    
    
def plot_pareto_tmp2(f1,pla,ax,title=None,color=False,y=None):
    legend = True
    for i in range(len(f1)):
        color_expl = colors[i]
        for j in range(len(f1[0])):
            color_gnn = colors[j]

            if f1[i][j] >= 0.5 and pla[i][j]>= 0.5:
               
            #if True: 
                if color:
                    ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN_bahousescolor[j],fontdict=dict(color=color_expl,size=10),weight="bold")
                        
                else:
                    ax.text(x=f1[i][j],y=pla[i][j]-0.02,s=GNN[j],fontdict=dict(color=color_expl,size=10),weight="bold")
                #ax.text(x=f1[i][j],y=pla[i][j],s=expl[i],fontdict=dict(color=color_gnn,size=10),weight="bold")


    ax.set_ylabel("Plausibility")
    ax.set_xlabel("Fidelity")

    #ax.set_xlim(0.7,1)
    #ax.set_ylim(0.7,1)
    ax.set_xlim(0.5,1)
    ax.set_ylim(0.5,1)
    
    if not title == None:
        ax.set_title(title)

    if legend:
        legend_elements_expl = []
        for i in range(len(expl)):
            ax.scatter(-1,-1,color=colors[i],label=expl[i])
        for i in range(len(expl)):
            legend_elements_expl.append(Line2D([0], [0], marker='o', color=colors[i], label=expl[i],))
    
        return legend_elements_expl
```
