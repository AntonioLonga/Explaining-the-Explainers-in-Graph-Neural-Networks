```python
import networkx as nx
import os

import matplotlib.pyplot as plt

def my_plot_edge(g,title=None,pos=None,ax=None):
    if not ax == None:
        ax.set_title(title)
    colors = list(nx.get_edge_attributes(g,"edge_imp").values())
    options = {
        "node_color":"#cccccc",
        "edge_color": colors,
        "width": 2,
        "edge_cmap": plt.cm.Reds,
        "with_labels": False,
        "node_size":25    }
    nx.draw(g,pos,ax,**options)
    
def my_plot(g,title=None,pos=None,ax=None):
    node_impo = nx.get_node_attributes(g,"node_imp")
    if not ax == None:
        ax.set_title(title)
    nx.draw(g,pos,ax,node_color=list(node_impo.values()), cmap=plt.cm.Reds,node_size=100,edgecolors="gray")
    
```

# example edge importance

```python
DATASET = "BA_2grid" # grid
MODEL = "HO"  # Graph Conv
EXPLAINER = "sal_edge" # GradExplEdge
MODE = "train" 
CLASS = "1"

if EXPLAINER in ["sal_edge","ig_edge","pgexpl","gnnexpl"]:
    FOLDER = "edge_imp"
else:
    FOLDER = "node_imp"
    
path = "GraphClassification/"+DATASET+"/"+MODEL+"/"+FOLDER+"/"+EXPLAINER+"/"+MODE+"/"+CLASS
graph_name = os.listdir(path)[1]
```

```python
# load the graph
g = nx.read_gpickle(path+"/"+graph_name)
```

```python
print("node impo: ",nx.get_node_attributes(g,"node_imp"))
print("edge impo: ",nx.get_edge_attributes(g,"edge_imp"))
```

```python
pos = nx.spring_layout(g,seed=11)
my_plot_edge(g,pos=pos)
```
