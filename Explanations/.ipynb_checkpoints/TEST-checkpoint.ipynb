```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
seed = 10
np.random.seed(seed)

def generate_star_network_BA(nb_nodesBA,nb_stars,nb_rewire=10,small_star=10,big_star=16):
    
    g = nx.fast_gnp_random_graph(nb_nodesBA,0.4)
    star_size = np.arange(small_star,big_star)
    for i in range(nb_stars):
        size = np.random.choice(star_size)
        star = nx.star_graph(size)
        g = nx.union(g,star,rename=("a","b"))

    mapping = dict()
    c = 0
    for n in g.nodes():
        mapping[n] = c
        c = c + 1
    g = nx.relabel_nodes(g,mapping)
    
    for i in range(nb_rewire):
        a,b = np.random.choice(g.nodes(),2,replace=False)
        if not g.has_edge(a,b):
            g.add_edge(a,b)
    return g


def gen_stars_BA_3class():
    graphs = []
    labels = []
    for i in range(500):

        g1 = generate_star_network_BA(nb_nodesBA = np.random.randint(20,25),
                                      nb_stars = 2,
                                      nb_rewire=10,small_star=10,big_star=16)
        
        g2 = generate_star_network_BA(nb_nodesBA = np.random.randint(15,20),
                                      nb_stars = 4,
                                      nb_rewire=10,small_star=10,big_star=16)
        
        g3 = generate_star_network_BA(nb_nodesBA = np.random.randint(10,15),
                                      nb_stars = 6,
                                      nb_rewire=10,small_star=10,big_star=16)
        
        if nx.is_connected(g1):
            if nx.is_connected(g2):
                if nx.is_connected(g3):
                    labels.append(0)
                    labels.append(1)
                    labels.append(2)
                    graphs.append(g1)
                    graphs.append(g2)
                    graphs.append(g3)
    
    return graphs,labels

graphs,labels = gen_stars_BA_3class()
len(graphs) 
```

```python
from typing import Optional, Callable

import torch
from torch_geometric.data import InMemoryDataset, Data
from torch_geometric.utils import  from_networkx
import random
import torch_geometric.transforms as T



class b2(InMemoryDataset):

    def __init__(self,graphs,labels,transform: Optional[Callable] = None,pre_filter: Optional[Callable] = None):
        super().__init__('.',transform,pre_filter)
        random.seed(10)


        #num_nodes = len(adjs[0][0])
        #self.num_nodes = num_nodes
        data_list = []
        for i in range(len(graphs)):
            num_nodes = len(graphs[i].nodes())
            label = labels[i]
            if random.random() <= 0.5:
                expl_mask = torch.zeros(num_nodes, dtype=torch.bool)
            else:
                expl_mask = torch.ones(num_nodes, dtype=torch.bool)
                
            data = from_networkx(graphs[i]) #to make edge_index und[feas]*10irected
            feas = np.array([1]*10)/10
            feas = [feas]*num_nodes
            data = Data(x=torch.tensor(feas), edge_index=data.edge_index, y=label,expl_mask=expl_mask)
            data_list.append(data)

        self.data, self.slices = self.collate(data_list)
        
        
dataset = b2(graphs,labels)
```

```python
from torch_geometric.utils import to_networkx

train_data = dataset[:int(len(dataset)*0.8)]
test_data = dataset[int(len(dataset)*0.8):]


from torch_geometric.loader import DataLoader
import torch
import torch.nn.functional as F
from torch.nn import BatchNorm1d, Linear
from torch_geometric.nn import GCNConv, global_mean_pool, GATConv, SAGEConv


train_loader = DataLoader(train_data,batch_size=128)
test_loader = DataLoader(test_data,batch_size=128)



class Net(torch.nn.Module):
    def __init__(self, num_features, num_classes):
        super().__init__()
        self.conv1 = GCNConv(num_features, 30)
        self.conv2 = GCNConv(30, 30)
        self.conv3 = GCNConv(30, 30)
        self.conv4 = GCNConv(30, 10)
        
        self.lin = Linear(10, num_classes)

    def forward(self, x, edge_index, batch,edge_mask=None):
        x = self.get_emb(x, edge_index,edge_mask)

        x = global_mean_pool(x, batch)
        x = self.lin(x)
        return F.log_softmax(x, dim=-1)

    def get_emb(self, x, edge_index,edge_mask):
        x = F.relu(self.conv1(x, edge_index,edge_mask))
        x = F.relu(self.conv2(x, edge_index,edge_mask))
        x = F.relu(self.conv3(x, edge_index,edge_mask))
        x = self.conv4(x, edge_index)
        return x
    
device = "cpu"
dataset.num_classes
```

```python
model = Net(10,dataset.num_classes).to(device).double()
optimizer = torch.optim.Adam(model.parameters(),lr=0.005) #,lr=0.0005) 


def train(train_loader):   
    model.train()
    optimizer.zero_grad()

    total_loss = 0
    for data in train_loader:
        data = data.to(device)
        optimizer.zero_grad()
        output = model(data.x, data.edge_index, data.batch)
        loss = F.nll_loss(output, data.y)
        loss.backward()
        optimizer.step()
        total_loss += float(loss) * data.num_graphs
    return total_loss / len(train_loader.dataset)

@torch.no_grad()
def test(loader):
    model.eval()

    total_correct = 0
    total_loss = 0
    for data in loader:
        data = data.to(device)
        out = model(data.x, data.edge_index, data.batch)
        total_correct += int((out.argmax(-1) == data.y).sum())

        loss = F.nll_loss(out, data.y)
        total_loss += float(loss) * data.num_graphs

    return total_correct / len(loader.dataset),total_loss / len(train_loader.dataset)


for epoch in range(1, 101): #400
    loss = train(train_loader)
    train_acc,train_loss = test(train_loader)
    test_acc,test_loss = test(test_loader)
    if epoch % 5 == 0:
        print(f'Epoch: {epoch:03d}, Loss: {loss:.3f}, Test Loss: {test_loss:.3f}, Train Acc: {train_acc:.3f} '
        f'Test Acc: {test_acc:.3f}')
```

```python
def ig_edge(model,data):
    def model_forward(edge_mask, data):
        batch = torch.zeros(data.x.shape[0], dtype=int).to(device)
        out = model(data.x, data.edge_index, batch, edge_mask)
        return out

    model = model.to("cpu")
    data = data.to("cpu")
    
    input_mask = torch.ones(data.edge_index.shape[1]).requires_grad_(True).to(device)
    pred = model(data.x, data.edge_index,data.batch).argmax(dim=-1)
    target = pred.cpu().detach().numpy().item()

    ig = IntegratedGradients(model_forward)
    mask = ig.attribute(input_mask, target=target,additional_forward_args=(data,),internal_batch_size=data.edge_index.shape[1])

    edge_mask = np.abs(mask.cpu().detach().numpy())
    if edge_mask.max() > 0:  # avoid division by zero
        edge_mask = edge_mask / edge_mask.max()

    data.edge_imp = edge_mask
    g = to_networkx(data,edge_attrs=["edge_imp"])
    g = nx.Graph(g)
    
    return g

def sal_edge(model,data):
    def model_forward(edge_mask, data):
        batch = torch.zeros(data.x.shape[0], dtype=int).to(device)
        out = model(data.x, data.edge_index, batch, edge_mask)
        return out

    model = model.to("cpu")
    data = data.to("cpu")
    
    input_mask = torch.ones(data.edge_index.shape[1]).requires_grad_(True).to(device)
    pred = model(data.x, data.edge_index,data.batch).argmax(dim=-1)
    target = pred.cpu().detach().numpy().item()

    saliency = Saliency(model_forward)
    mask_s = saliency.attribute(input_mask, target=target,additional_forward_args=(data,))

    edge_mask_s = np.abs(mask_s.cpu().detach().numpy())
    if edge_mask_s.max() > 0:  # avoid division by zero
        edge_mask_s = edge_mask_s / edge_mask_s.max()
    edge_mask = edge_mask_s


    data.edge_imp = edge_mask
    g = to_networkx(data,edge_attrs=["edge_imp"])
    g = nx.Graph(g)
    
    return g

def gnnexpl(model,data):
    model = model.to("cpu")
    data = data.to("cpu")
    
    explainer = GNNExplainer(model,num_hops=None,epochs=500,return_type='log_prob')
    node_feat_mask, edge_mask = explainer.explain_graph(data.x, data.edge_index)
    data.edge_imp = edge_mask
    g = to_networkx(data,edge_attrs=["edge_imp"])
    g = nx.Graph(g)
    
    return g


def normalize_weights(g):
    weights = nx.get_node_attributes(g,"node_imp")
    w_list = np.array(list(weights.values()))
    minw = min(w_list)
    maxw = max(w_list-minw)

    for k in weights:
        t = (weights[k]-minw)
        weights[k] = t/maxw

    for n in g.nodes():
        g.nodes()[n]["node_imp"] = weights[n]
    return g

def cam(model,data):
    model = model.to("cpu")
    data = data.to("cpu")
    cam = CAM(model)
    a = cam.get_explanation_graph(data.x,data.edge_index,data.y,forward_kwargs= {"batch":data.batch})
    g = a.graph_to_networkx(to_undirected=True)
    return normalize_weights(g)


def grad_cam(model,data,criterion=F.nll_loss):
    model = model.to("cpu")
    data = data.to("cpu")
    gcam =GradCAM(model, criterion=criterion)
    a = gcam.get_explanation_graph(data.x,data.edge_index,data.y,forward_kwargs= {"batch":data.batch})
    g = a.graph_to_networkx(to_undirected=True)
    return normalize_weights(g)


def guided_bp(model,data,criterion=F.nll_loss):
    model = model.to("cpu")
    data = data.to("cpu")
    guidedBP = GuidedBP(model,criterion = criterion)
    a = guidedBP.get_explanation_graph(data.x,data.y,data.edge_index,forward_kwargs= {"batch":data.batch})
    
    g = a.graph_to_networkx(to_undirected=True)
    return normalize_weights(g)

def ig_node(model,data,criterion=F.nll_loss):
    model = model.to("cpu")
    data = data.to("cpu")
    ig = IntegratedGradExplainer(model,criterion=criterion)
    a = ig.get_explanation_graph(data.edge_index,data.x,data.y,forward_kwargs= {"batch":data.batch})
    
    g = a.graph_to_networkx(to_undirected=True)
    return normalize_weights(g)

def grad_exp(model,data,criterion=F.nll_loss):
    model = model.to("cpu")
    data = data.to("cpu")
    grade = GradExplainer(model,criterion=criterion)
    a = grade.get_explanation_graph(data.x,data.edge_index,data.y,forward_kwargs= {"batch":data.batch})
    g = a.graph_to_networkx(to_undirected=True)
    return normalize_weights(g)

def pgmexplainer(model,data,perturb_mode = "mean"):
    model = model.to("cpu")
    data = data.to("cpu")
    pgme = PGMExplainer(model,explain_graph=True,perturb_mode=perturb_mode)
    a = pgme.get_explanation_graph(data.x,data.edge_index,data.y,forward_kwargs= {"batch":data.batch})
    g = a.graph_to_networkx(to_undirected=True)
    
    return normalize_weights(g)
```

```python
def subplot_expl_node3(expl,model,datas,poss):
    fig, axs = plt.subplots(1,3,figsize=(15,5))
    if poss == None:
        poss = []
    for i in range(3):
        g = expl(model,datas[i])
        if i == 2:
            plot_exp_node(g,axs[i],poss[i],colorbar=True)
        else:
            plot_exp_node(g,axs[i],poss[i])
            

def subplot_expl_edge3(expl,model,datas,poss):
    fig, axs = plt.subplots(1,3,figsize=(15,5))
    if poss == None:
        poss = []
    for i in range(3):
        g = expl(model,datas[i])
        if i == 2:
            plot_exp_edge(g,axs[i],poss[i],colorbar=True)
        else:
            plot_exp_edge(g,axs[i],poss[i])
            
def plot_exp_node(g,ax,pos=None,colorbar=False):
    node_impo = nx.get_node_attributes(g,"node_imp")
    nx.draw(g,pos,ax=ax,node_color=list(node_impo.values()), cmap=plt.cm.Reds,node_size=100)
    if colorbar:
        sm = plt.cm.ScalarMappable(cmap=plt.cm.Reds)
        sm._A = []
        plt.colorbar(sm)
        
def plot_exp_edge(g,ax,pos=None,colorbar=False):
    edge_impo = nx.get_edge_attributes(g,"edge_imp")
    nx.draw(g,pos,ax=ax,edge_color=list(edge_impo.values()), edge_cmap=plt.cm.Reds,node_size=1,width=2)
    if colorbar:
        sm = plt.cm.ScalarMappable(cmap=plt.cm.Reds)
        sm._A = []
        plt.colorbar(sm)
```
